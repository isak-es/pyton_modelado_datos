# 2. Ventas por categoría

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Diccionario con nombres de tiendas y URLs de los archivos CSV
urls = {
    "Tienda_1": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv",
    "Tienda_2": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv",
    "Tienda_3": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv",
    "Tienda_4": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"
}

#DataFrame vacío para almacenar ventas por categoría
dataframes = []

#archivos CSV y calcular ventas por categoría
for tienda, url in urls.items():
    try:
        df = pd.read_csv(url)

        # Verificar que la columna "Categoría del Producto" existe
        if "Categoría del Producto" in df.columns:
            ventas_por_categoria = df["Categoría del Producto"].value_counts().reset_index()
            ventas_por_categoria.columns = ["Categoría del Producto", "Cantidad Vendida"]
            ventas_por_categoria["Tienda"] = tienda  # Agregar la tienda
            dataframes.append(ventas_por_categoria)
        else:
            print(f"Advertencia: La columna 'Categoría del Producto' no fue encontrada en {tienda}")

    except Exception as e:
        print(f"Error al leer {tienda}: {e}")

# Verificar si hay datos antes de concatenar
if dataframes:
    df_final = pd.concat(dataframes, ignore_index=True)
else:
    print("No se encontraron datos para concatenar.")

# Visualización de ventas por categoría
plt.figure(figsize=(12, 6))
sns.barplot(x="Categoría del Producto", y="Cantidad Vendida", hue="Tienda", data=df_final, palette="Blues")



plt.xlabel("Categorías", fontsize=12)
plt.ylabel("Cantidad Vendida", fontsize=12)
plt.title("Ventas por Categoría en Cada Tienda", fontsize=14)
plt.xticks(rotation=45)
plt.legend(title="Tiendas")
plt.grid(axis='y', linestyle='--', alpha=0.6)

# Mostrar gráfico
plt.show()



```

```python
import pandas as pd

# Diccionario con nombres de tiendas y URLs de los archivos CSV
urls = {
    "Tienda_1": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv",
    "Tienda_2": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv",
    "Tienda_3": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv",
    "Tienda_4": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"
}

# Lista para almacenar la valoración media de cada tienda
valores = []

#archivos CSV y calcular la valoración promedio
for tienda, url in urls.items():
    try:
        df = pd.read_csv(url)

        # Verificar que la columna "Calificación" existe
        if "Calificación" in df.columns:
            valoracion_media = df["Calificación"].mean()
            valores.append({"Tienda": tienda, "Valoración Media": round(valoracion_media, 2)})  # Redondear a 2 decimales
        else:
            print(f"Advertencia: La columna 'Calificación' no fue encontrada en {tienda}")

    except Exception as e:
        print(f"Error al leer {tienda}: {e}")

# Convertir lista en DataFrame
df_valoraciones = pd.DataFrame(valores)

# Mostrar los resultados
print("Valoración media por tienda:")
print(df_valoraciones)
```

# 4. Productos más y menos vendidos

```python
import pandas as pd

# Diccionario con nombres de tiendas y URLs de los archivos CSV
urls = {
    "Tienda_1": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv",
    "Tienda_2": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv",
    "Tienda_3": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv",
    "Tienda_4": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"
}

# Iterar sobre cada tienda y calcular los productos más vendidos y menos vendidos
for tienda, url in urls.items():
    try:
        df = pd.read_csv(url)

        # Verificar que las columnas "Producto" y "Cantidad de cuotas" existen
        if "Producto" in df.columns and "Cantidad de cuotas" in df.columns:
            # Agrupar por producto y sumar la cantidad de cuotas (usándolo como cantidad vendida)
            productos_ordenados = df.groupby("Producto")["Cantidad de cuotas"].sum().reset_index().sort_values(by="Cantidad de cuotas", ascending=False)

            # Obtener el producto más vendido y menos vendido
            mas_vendido = productos_ordenados.iloc[0]  # Producto más vendido
            menos_vendido = productos_ordenados.iloc[-1]  # Producto menos vendido

            # Imprimir resultados
            print(f"Resultados para {tienda}:")
            print(f"Producto más vendido: {mas_vendido['Producto']} ({mas_vendido['Cantidad de cuotas']} unidades)")
            print(f"Producto menos vendido: {menos_vendido['Producto']} ({menos_vendido['Cantidad de cuotas']} unidades)")
            print("-" * 40)

        else:
            print(f"Advertencia: No se encontraron las columnas necesarias en {tienda}")

    except Exception as e:
        print(f"Error al leer {tienda}: {e}")


```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Diccionario con nombres de tiendas y URLs de los archivos CSV
urls = {
    "Tienda_1": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv",
    "Tienda_2": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv",
    "Tienda_3": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv",
    "Tienda_4": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"
}

# Lista para almacenar los datos
data = []

#archivos CSV y calcular los productos más vendidos y menos vendidos
for tienda, url in urls.items():
    try:
        df = pd.read_csv(url)

        # Verificar que las columnas "Producto" y "Cantidad de cuotas" existen
        if "Producto" in df.columns and "Cantidad de cuotas" in df.columns:
            # Agrupar por producto y sumar la cantidad de cuotas (usándolo como cantidad vendida)
            productos_ordenados = df.groupby("Producto")["Cantidad de cuotas"].sum().reset_index().sort_values(by="Cantidad de cuotas", ascending=False)

            # Obtener el producto más vendido y menos vendido
            mas_vendido = productos_ordenados.iloc[0]
            menos_vendido = productos_ordenados.iloc[-1]

            # Almacenar datos
            data.append({"Tienda": tienda, "Producto": mas_vendido["Producto"], "Cantidad Vendida": mas_vendido["Cantidad de cuotas"], "Tipo": "Más Vendido"})
            data.append({"Tienda": tienda, "Producto": menos_vendido["Producto"], "Cantidad Vendida": menos_vendido["Cantidad de cuotas"], "Tipo": "Menos Vendido"})

        else:
            print(f"Advertencia: No se encontraron las columnas necesarias en {tienda}")

    except Exception as e:
        print(f"Error al leer {tienda}: {e}")

# Convertir lista en DataFrame
df_resultados = pd.DataFrame(data)

# 🎨 **Gráfico de barras con nombres de productos**
plt.figure(figsize=(12, 6))
ax = sns.barplot(x="Tienda", y="Cantidad Vendida", hue="Tipo", data=df_resultados, palette=["blue", "red"])

# Agregar nombres de productos sobre las barras
for p, label in zip(ax.patches, df_resultados["Producto"]):
    ax.text(p.get_x() + p.get_width() / 2, p.get_height(), label, ha="center", va="bottom", fontsize=10, color="black")

#gráfico
plt.xlabel("Tiendas", fontsize=12)
plt.ylabel("Cantidad Vendida", fontsize=12)
plt.title("Productos más vendidos y menos vendidos por tienda", fontsize=14)
plt.legend(title="Tipo")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.xticks(rotation=45)

# Mostrar gráfico
plt.show()

```

# 5. Envío promedio por tienda

```python
import pandas as pd

# Diccionario con nombres de tiendas y URLs de los archivos CSV
urls = {
    "Tienda_1": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv",
    "Tienda_2": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv",
    "Tienda_3": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv",
    "Tienda_4": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"
}

# Lista para almacenar el costo de envío promedio de cada tienda
envios = []

# archivos CSV y calcular el costo promedio de envío
for tienda, url in urls.items():
    try:
        df = pd.read_csv(url)

        # Verificar que la columna "Costo de envío" existe
        if "Costo de envío" in df.columns:
            envio_promedio = df["Costo de envío"].mean()
            envios.append({"Tienda": tienda, "Envío Promedio": round(envio_promedio, 2)})  # Redondear a 2 decimales
        else:
            print(f"Advertencia: La columna 'Costo de envío' no fue encontrada en {tienda}")

    except Exception as e:
        print(f"Error al leer {tienda}: {e}")

# Convertir lista en DataFrame
df_envios = pd.DataFrame(envios)

# Mostrar los resultados
print("Envío promedio por tienda:")
print(df_envios)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Diccionario con nombres de tiendas y URLs de los archivos CSV
urls = {
    "Tienda_1": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv",
    "Tienda_2": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv",
    "Tienda_3": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv",
    "Tienda_4": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"
}

# Lista para almacenar el envío promedio por tienda
envios = []

#archivos CSV y calcular el número promedio de envíos
for tienda, url in urls.items():
    try:
        df = pd.read_csv(url)

        # Verificar que la columna "Cantidad de cuotas" existe
        if "Cantidad de cuotas" in df.columns:
            envio_promedio = df["Cantidad de cuotas"].mean()
            envios.append({"Tienda": tienda, "Envío Promedio": round(envio_promedio, 2)})  # Redondear a 2 decimales
        else:
            print(f"Advertencia: La columna 'Cantidad de cuotas' no fue encontrada en {tienda}")

    except Exception as e:
        print(f"Error al leer {tienda}: {e}")

# Convertir lista en DataFrame
df_envios = pd.DataFrame(envios)

# Generación del gráfico de torta
plt.figure(figsize=(8, 8))
plt.pie(df_envios["Envío Promedio"], labels=df_envios["Tienda"], autopct="%1.1f%%", colors=["blue", "orange", "green", "red"])

# Personalizar gráfico
plt.title("Número Promedio de Envíos por Tienda")
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Diccionario con nombres de tiendas y URLs de los archivos CSV
urls = {
    "Tienda_1": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/main/base-de-datos-challenge1-latam/tienda_1.csv",
    "Tienda_2": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/main/base-de-datos-challenge1-latam/tienda_2.csv",
    "Tienda_3": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/main/base-de-datos-challenge1-latam/tienda_3.csv",
    "Tienda_4": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/main/base-de-datos-challenge1-latam/tienda_4.csv"
}

# Crear figura
plt.figure(figsize=(10, 6))

# Leer archivos y graficar cada tienda
for tienda, url in urls.items():
    try:
        df = pd.read_csv(url)

        # Verificar que las columnas "lat" y "lon" existen
        if "lat" in df.columns and "lon" in df.columns:
            plt.scatter(df["lon"], df["lat"], label=tienda, alpha=0.6)

    except Exception as e:
        print(f"Error al leer {tienda}: {e}")

#gráfico
plt.xlabel("Longitud")
plt.ylabel("Latitud")
plt.title("Distribución Geográfica de Ventas por Tienda")
plt.legend()
plt.grid()

# Mostrar gráfico
plt.show()

```

```python

import seaborn as sns

plt.figure(figsize=(10, 6))

# Unir datos de todas las tiendas
dataframes = []

for tienda, url in urls.items():
    try:
        df = pd.read_csv(url)
        if "lat" in df.columns and "lon" in df.columns:
            dataframes.append(df[["lat", "lon"]])

    except Exception as e:
        print(f"Error al leer {tienda}: {e}")

# Concatenar los datos si hay archivos disponibles
if dataframes:
    df_total = pd.concat(dataframes)

    # Crear mapa de calor
    sns.kdeplot(x=df_total["lon"], y=df_total["lat"], cmap="Reds", fill=True)

    plt.xlabel("Longitud")
    plt.ylabel("Latitud")
    plt.title("Mapa de calor de ventas por ubicación")
    plt.grid()

    plt.show()
else:
    print("No se encontraron datos válidos para generar el mapa de calor.")
```
